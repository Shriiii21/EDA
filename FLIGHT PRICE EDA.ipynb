```python
#feature engineering
df['DATE']=df['Date_of_Journey'].str.split('/').str[0]
df['MONTH']=df['Date_of_Journey'].str.split('/').str[1]
df['YEAR']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.info()
```

```python
df['DATE']=df['DATE'].astype(int)
df['MONTH']=df['MONTH'].astype(int)
df['YEAR']=df['YEAR'].astype(int)
```

```python
df.info()
```

```python
#drop date of journey colm
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python

df['ARRIVAL HOUR']=df['Arrival_Time'].str.split(':').str[0]
df['ARRIVAL MINUTE']=df['Arrival_Time'].str.split(':').str[1]
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()

encoder.fit_transform(df[['Airline','Source','Destination']]).toarray()
```

```python
encoded_data=pd.DataFrame(encoder.fit_transform(df[['Airline','Source','Destination']]).toarray(),columns=encoder.get_feature_names_out())
```

```python
#  Drop original categorical columns
df = df.drop(['Airline', 'Source', 'Destination'], axis=1)
df = pd.concat([df, encoded_data], axis=1)
```
